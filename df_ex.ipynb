```python
import pandas as pd
```

```python
alist = ['Mouse', 'Giraffe', 'Tiger', 'Dog', 'Hippopotamus', 'Moose', 'Bear']
animals = pd.Series(alist)
animals
```

```python
animals[2]
```

```python
animals.iloc[2]
```

```python
animals.loc[2]
```

```python
s1 = pd.Series(alist, index=[0, 1, 2, 3, 4, 5, 6])
s1.iloc[5]
```

```python
s1.loc[2]
```

```python
s2 = pd.Series(alist, index=['a', 'b', 'c', 'd', 'e', 'f', 'g'])
s2
```

```python
s2['c']
```

```python
s2.iloc[2]
```

```python
names = ['Chris', 'Kevin', 'Bob', 'Vinod', 'Daniel', 'Fil', 'Stephanie']
owners = pd.Series(names)
owners
```

```python
frame = {'animals': animals, 'owners': owners}

df = pd.DataFrame(frame)

df
```

```python
df.columns
```

```python
df.index
```

```python
type(df['animals'])
```

```python
df.iloc[0]
```

```python
df.iloc[0][1]
```

```python
df.owners
```

```python
df['owners']
```

```python
# copy an existing DataFrame to create a new dataFrame
copy_df = df.copy()
copy_df
```

```python
# drop columns

df2 = df.drop(['owners'], axis = 1)

df2
```

```python
# drop columns

df3 = df.drop(columns = ['owners'])

df3
```

```python
# drop rows

df4 = df.drop([0,1])
df4
```

```python
midx = pd.MultiIndex(levels=[['lama', 'cow', 'falcon'],
                             ['speed', 'weight']],
                     codes=[[0, 0, 1, 1, 2, 2],
                            [0, 1, 0, 1, 0, 1]])

midx
```

```python
df = pd.DataFrame(index=midx, columns=['big', 'small'],
                   data=[[45, 30], [200, 100], [30, 20], [250, 150], [320, 250],[1, 0.8]])

df
```

```python
df.drop(index='cow', columns='small')
```

```python
df.drop(index='weight', level=1)
```
